## 1. Introduction to the Seaborn Library

### 1.1. Background

Created by Michael Waskom in 2014, [Seaborn](https://seaborn.pydata.org/) is an open-source Python library dedicated to statistical data visualisation. It is built on top of [matplotlib](https://matplotlib.org/), and closely integrated with [pandas](https://pandas.pydata.org/) data structures. As Seaborn sits on top of matplotlib and extends it, its learning curve is much more gradual for users who start with matplotlib and want to expand their plotting arsenal as compared to other visualization packages.


### 1.2. Installation

There are several ways to install Seaborn's latest release:  
- Seaborn can be installed using pip: *pip install seaborn*  
- It can be installed using conda: *conda install seaborn*  
- It can also be installed using pip to get the development version directly from github: *pip install git+https://github.com/mwaskom/seaborn.git*  


### 1.3 Dependencies

Seaborn is supported on Python 2.7 or 3.5+.  
Main dependencies:

- numpy (>= 1.9.3)
- scipy (>= 0.14.0)
- matplotlib (>= 1.4.3)
- pandas (>= 0.15.2)

Recommended dependency:

- statsmodels (>= 0.5.0)

### 1.4. Using Seaborn for multivariate datasets

Seaborn is a great choice for visualizing multivariate datasets. Its functional structure allows for easy plotting of categorical variables to show observations or aggregate statistics. It also has many options for visualizing bivariate and univariate distributions and for comparing them between subsets of data. Lastly, Seaborn has high-level abstractions for structuring multi-plot grids that let users build complex visualizations extremely quickly.

Seaborn draws plots using matplotlib behind the scene. Many tasks can be accomplished with only Seaborn functions, but further customization might require using matplotlib directly, which is probably Seaborn's most glaring limitation. For interactive work, it’s recommended to use a Jupyter/IPython interface in matplotlib mode, or else you’ll have to call matplotlib.pyplot.show whenever you want to see the plot.

In the next section, we will explore Seaborn's capabilities listed above through visualizing the relationship between two and more variables.


## 2. Seaborn In Action

### 2.1. The Dataset

For illustration, we will be using the College Scorecard data, produced by the US Department of Education. The College Scorecard is designed to increase transparency for students and parents, allowing them to find schools that best fit what they are looking for while ensuring a broad and flexible definition of elements that contribute to "fit".

To understand more on what each of the variable means, refer to College Scorecard Data Dictionary [here.](https://collegescorecard.ed.gov/data/documentation/)

We'll start by importing the necessary libraries:
- Numpy and Pandas for data processing
- Seaborn and Matplotlib for data visualization

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the College Scorecard dataset and store it as a Pandas dataframe
df = pd.read_csv('CollegeScorecard.csv', encoding = "ISO-8859-1")
df.head()
```

```python
# The CONTROL column captures the type of institutions as numeric values. 
# We will convert them to categorical for later use.

df['CONTROL'] = df['CONTROL'].replace({1: 'Public',
                                      2: 'Private nonprofit',
                                      3: 'Private for-profit'})
```

### 2.2. Relationship between Admission Rate and SAT Score

Bivariate data arises when every "unit of analysis", in this case every student, is assessed with respect to two traits. We will use Seaborn to visualizate how Average SAT Score and Admission Rate variables relate to each other. Do Top universities eg Ivy League with high SAT Avg Score will perhaps also have a lower admission rate, making them more competitive?

**Scatterplot** a type of statistical visualization that displays the relationship between two numerical variables. Each member of the dataset gets plotted as a point whose (x, y) coordinates relates to its values for the two variables. As a result, scatterplots help to visualize patterns or relationships between the variables, specifically correlation (positive if one variable increases as the other increases and negative vice versa). From scatterplots, we can also tell whether the relationship between two variables are linear or not or if data is divided into different clusters.

#### 2.2.1 Relplot

This function provides access to several different axes-level functions that show the relationship between two variables. The kind parameter selects the underlying axes-level function to use, with 'scatter' being the default.

```python
sns.relplot(x='ADM_RATE', y='SAT_AVG', data=df, kind='scatter')
```

While the points are plotted in two dimensions, another dimension can be added to the plot by coloring the points according to a third variable. In seaborn, this is referred to as using a “hue semantic”, because the color of the point gains meaning.

```python
sns.relplot(x='ADM_RATE', y='SAT_AVG', hue='CONTROL', data=df, kind='scatter')
```

#### 2.2.2 Regplot

An alternative to relplot is regplot. Regplot also plots the data in a similar fashion, but includes a linear regression model fit. This obviously would not work so well for non-linear relationships; nevertheless, you can toggle the best fit line by changing the fit_reg parameter. In this case, we'll set it to False to not display the best fit line.

```python
sns.regplot(x='ADM_RATE', y='SAT_AVG', data=df, fit_reg=False)
```

#### 2.2.3 Scatterplot

Next up is scatterplot. As the name suggests, scatterplot function generates a scatterplot, but in Seaborn, it is possible to include semantic groupings. This is a big improvement over matplotlib's scatter function. 

The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters.

```python
sns.scatterplot(x='ADM_RATE', y='SAT_AVG', hue='CONTROL', data=df)
```

### 2.3 Improving and enhancing Seaborn scatterplots

#### 2.3.1 Enhancing scatterplots

We have illustrated that there are more than one way to draw scatterplots using Seaborn. All three visualizations are more or less the same and they display a somewhat negative correlation between the Avg. SAT scores and Admission Rates, suggesting that as Avg. SAT scores increase, Admission Rates decrease; however, this is far from causality.

In this section, we will improve the existing scatterplots to better visualize the relationship.

First and foremost, in a scatterplot with more than around 100 points, **"overplotting"** becomes an issue.  This means that many points fall on top of each other in the plot, which obscures relationships in the middle of the distribution and over-emphasizes the extremes.  One way to mitigate overplotting is to use an "alpha" channel to make the points semi-transparent.

```python
sns.regplot(x='ADM_RATE', y='SAT_AVG', data=df, fit_reg=True, scatter_kws={"alpha": 0.2})
```

Another way to avoid overplotting is to make a plot of the "density" of points.  In the plots below, darker colors indicate where a greater number of points fall.  The two plot margins show the densities for the Avg SAT scores and Admission Rates separately, while the plot in the center shows their density jointly. We use Seaborn's jointplot function for this.

```python
sns.jointplot(x='ADM_RATE', y='SAT_AVG', kind='kde', data=df)
```

```python
# It's of course also possible to do a jointplot with scatterplot instead of density plot. It's in fact the default.
sns.jointplot(x='ADM_RATE', y='SAT_AVG', data=df)
```

Next, it would be helpful to include an annotation showing the actual correlation value, especially in this case where it is not exactly clear how strongly negative the correlation is. To do this, we will need to import the pearsonr function from scipy stats, and include it in an annotate function.

```python
import scipy.stats as stats
from warnings import filterwarnings
filterwarnings('ignore')

sns.jointplot(x='ADM_RATE', y='SAT_AVG', kind='kde', data=df).annotate(stats.pearsonr)
```

Now it's clear that the correlation is only slightly negative. 

Lastly, if we look at the scatterplots created using relplot and scatterplot function above, Seaborn allows the inclusion of groupings. But because the number of data points is too big, it is difficult to ascertain the relationship between the 2 variables for each group category. This is where Seaborn provides a great function - FacetGrid. 

FacetGrid lets users quickly and easily create a series of plots, one for each specified categorical value. In this case, we can plot 3 separate scatterplots for each institution type instead of having everything on the same plot.

```python
sns.FacetGrid(df, col='CONTROL').map(plt.scatter, 'ADM_RATE', 'SAT_AVG', alpha=.2).add_legend()
```

#### 2.3.2 Customizing aesthetic features

Seaborn has a rich library of aesthetic customizations, and while we cannot touch on every single one of them, I encourage that you check out the [Seaborn documentation](https://seaborn.pydata.org/tutorial.html) for a more comprehensive list of supported aesthetic features.

**Setting figure style and scale**

Being built on top of matplotlib, seaborn allows you to revert to the matplotlib default theme if that's the preference. Seaborn itself has five preset themes - *darkgrid*, *whitegrid*, *dark*, *white*, and *ticks*. Each theme is suited to different application and visualization purposes.  
The Seaborn default is darkgrid, but when visualizing a lot of data elements eg group of boxplots, it's better to switch over to whitegrid. Or you can remove the grid altogether if the visualization is built to emphasize data patterns.

```python
sns.set_style("dark")
```

The spines aren't really necessary, so we can remove them as well. You can also parse an *offset* parameter to push the spines away from the data.

```python
sns.despine()
```

To customize the seaborn styles, you can pass a dictionary of parameters to the rc argument of axes_style() and set_style(). Note that you can only override the parameters that are part of the style definition through this method. If you want to see what parameters are included, you can just call the function with no arguments, which will return the current settings. 

Seaborn also have preset scaling modes for a quick way to adjust to the scale of the figures through calling set_context(). The 4 present contexts are *paper*, *notebook*, *talk*, and *poster*. The *notebook* style is the default.

```python
# Let's try to plot the FacetGrid again

sns.set_style("white")
sns.despine()
sns.set_context("paper")
sns.FacetGrid(df, col='CONTROL').map(plt.scatter, 'ADM_RATE', 'SAT_AVG', alpha=.2).add_legend()
```

Seaborn makes it easy to select and use color palettes that are suited to the kind of data you are working with and the goals you have in visualizing it. The available color palettes are *deep, muted, bright, pastel, dark, colorblind*. Seaborn's set_color() function will set the default color palette for all plots. 

To change the color of our data points, use the palette parameter. In addition to the 6 built-in color palettes above, the palette parameter will accept whatever Seaborn's color_palette() function accepts. More information available [here.](https://seaborn.pydata.org/tutorial/color_palettes.html#palette-tutorial)

```python
# Let's try changing the color palette for our scatterplot to the built-in bright color palette

sns.scatterplot(x='ADM_RATE', y='SAT_AVG', hue='CONTROL', palette="bright", data=df)
```

```python
# We can also parse a list of hex codes

sns.scatterplot(x='ADM_RATE', y='SAT_AVG', hue='CONTROL', palette=('#2A3132','#763626', '#336B87'), data=df)
```

```python
# Or we can even parse a sequential color palette, albeit joinplot would automatically take care of this

sns.scatterplot(x='ADM_RATE', y='SAT_AVG', hue='CONTROL', palette="GnBu_d", data=df)
```

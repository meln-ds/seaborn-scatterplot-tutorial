```python
# Ok, third time's the charm!

%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('CollegeScorecard.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
sns.regplot(x='ADM_RATE', y='SAT_AVG', data=df, fit_reg=False)
```

```python
sns.regplot(x='ADM_RATE', y='SAT_AVG', data=df, fit_reg=True, scatter_kws={"alpha": 0.2})
```

```python
sns.jointplot(x='ADM_RATE', y='SAT_AVG', kind='kde', data=df)
```

```python
df['CONTROL'] = df['CONTROL'].replace({1: 'Public',
                                      2: 'Private nonprofit',
                                      3: 'Private for-profit'})
```

```python
sns.FacetGrid(df, col='CONTROL').map(plt.scatter, 'ADM_RATE', 'SAT_AVG', alpha=.2).add_legend()
```
